# 2. Pass / Fail Analyzer

Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]


```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for i in marks:
    if i >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("pass count is", pass_count)
print("fail count is", fail_count)
```

# 3. Simple Data Cleaner

Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]


```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_name = []

for name in names:
    clean = name.strip().lower()
    cleaned_name.append(clean)
print("cleaned name list is ", cleaned_name)
```

# 4. Message Length Analyzer

Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print("message", msg)
    print("length",length)
    
    if length > 10:
        print("this message is long")
    print()
```

# 5. Error Message Detector

Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1
print("the error message length is:", error_count)
```
