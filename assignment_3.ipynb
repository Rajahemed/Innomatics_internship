# Problem 1: Employee Performance Bonus Eligibility

```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Output
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", max_score, ")")
```

# Problem 2: Search Query Keyword Analysis

```python
# Problem 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Display words with frequency > 1
result = {}

for word, count in freq.items():
    if count > 1:
        result[word] = count

print(result)
```

# Problem 3: Sensor Data Validation

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem 4: Email Domain Usage Analysis

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)

print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

# Problem 5: Sales Spike Detection

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)

# Detect spikes (more than 30% above average)
threshold = avg_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print("Day", day, ":", value)
```

# Problem 6: Duplicate User ID Detection

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

for user, count in id_count.items():
    if count > 1:
        print(user, "→", count, "times")
```
