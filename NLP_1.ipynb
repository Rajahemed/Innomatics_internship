# 1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different tokens because NLP models are case-sensitive by default.
"Love" may appear at the beginning of a sentence, while "love" may appear in the middle. If we do not convert text to lowercase, the model will treat them as separate words, increasing vocabulary size and reducing model efficiency.
Therefore, we usually convert all text to lowercase during preprocessing.

# 2. What happens if stopwords are not removed?

# If stopwords are not removed:

The dataset will contain many common words like is, the, am, are, was.
These words do not add significant meaning.
They increase dataset size and processing time.
They may reduce model accuracy because meaningful words get less importance.

# 3. Two real-world scenarios where removing stopwords can be harmful

Removing stopwords is harmful in:

Sentiment Analysis
Example: "I am not happy"
If we remove "not", the sentence becomes "I am happy" → wrong meaning.
Chatbots / Question Answering Systems
Example: "What is the time now?"
Removing stopwords may remove important structure and make sentence unclear.

# 4. Difference between Stemming and Lemmatization

| Stemming                         | Lemmatization                         |
| -------------------------------- | ------------------------------------- |
| Cuts words to root form          | Converts word to meaningful base form |
| May produce non-dictionary words | Produces dictionary words             |
| Faster                           | Slower                                |
| Example: "running" → "run"       | "running" → "run"                     |
| Example: "studies" → "studi"     | "studies" → "study"                   |


# Task 2: Advanced Preprocessing Function

```python
import re
import string
from collections import Counter

# Exception words to keep even if length <= 2
EXCEPTION_WORDS = {"no", "not"}

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""
    
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # Remove emojis and special characters
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Handle repeated characters (soooo -> soo)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenization
    tokens = text.split()
    
    # Remove short tokens
    tokens = [word for word in tokens if len(word) > 2 or word in EXCEPTION_WORDS]
    
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
text = "I loooove NLP!!! Visit https://openai.com now"
tokens, clean_sentence = preprocess_text(text)

print("Tokens:", tokens)
print("Clean Sentence:", clean_sentence)
```

# Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sentence)
    print("-"*50)
```

# Task 4: Token Analytics

```python
def token_analytics(tokens):
    if len(tokens) == 0:
        return 0, 0, 0
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens
    
    return total_tokens, unique_tokens, round(avg_length, 2)


for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    total, unique, avg = token_analytics(tokens)
    
    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", avg)
```

# Task 5 — Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(freq.most_common()[-5:])
```

# Task 6 — Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)
print(output)
```
